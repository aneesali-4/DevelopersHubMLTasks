```python
#objective: exploring and visualizing the Iris dataset to understand feature distributions and relationships

#importing required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#loading iris dataset
df = sns.load_dataset('iris')
```

```python
#inspecting the shape, columns, and top rows of the dataset
print("Shape of dataset:", df.shape)
print("Columns:", df.columns.tolist())
df.head()
```

```python
#getting dataset info and statistical summary
df.info()
df.describe()
```

```python
#creating scatter plot matrix to show feature relationships
sns.pairplot(df, hue='species', diag_kind='hist')
plt.suptitle("scatter matrix of iris features by species", y=1.02)
plt.show()
```

```python
#plotting histograms for each feature
df.hist(edgecolor='black', figsize=(10, 8))
plt.suptitle("histograms of iris features")
plt.tight_layout()
plt.show()
```

```python
#creating box plots to detect outliers for each feature
plt.figure(figsize=(10, 6))
for i, col in enumerate(df.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.boxplot(data=df, x='species', y=col)
    plt.title(f'box plot of {col} by species')
plt.tight_layout()
plt.show()
```
